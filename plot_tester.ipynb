```python
import os
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.layouts import column
from bokeh.palettes import Turbo256
# CSV 파일 읽기

db = {}
data_fold = './Data_2/data_training/'
for filename in os.listdir(data_fold):
    if filename.endswith('.csv'):
        file_path = os.path.join(data_fold, filename)
        db[filename] = pd.read_csv(file_path)

p1 = figure(width=1000, height=1800, title="Simple line example", x_axis_label='x', y_axis_label='y')
for key in db.keys():
    x = [i for i in range(0,200)]
    p1.line(x, db[key]['V2'], legend_label=key, muted_alpha=0.0, line_width=2)
p1.legend.click_policy="mute"

p2 = figure(width=1000, height=1800, title="Simple line example", x_axis_label='x', y_axis_label='y')
for key in db.keys():
    x = [i for i in range(0,200)]
    p2.line(x, db[key]['V8'], legend_label=key, muted_alpha=0.0, line_width=2)
p2.legend.click_policy="mute"

p3 = figure(width=1000, height=800, title="Simple line example", x_axis_label='x', y_axis_label='y')
for key in db.keys():
    x = [i for i in range(0,200)]
    p3.line(x, db[key]['V10'], legend_label=key, muted_alpha=0.0, line_width=2)
p3.legend.click_policy="mute"

show(column(p1, p2, p3))


```
