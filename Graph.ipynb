```python
import os
import matplotlib.pyplot as plt
import pandas as pd


file1 = './Data_3/data_training/FAMP_C1U_2.0.csv'
df1 = pd.read_csv(file1)
x1_v2 = df1['V2']
x1_v8 = df1['V8']
x1_v10 = df1['V10']
time = df1['TIME1']

file2 = './Data_3/data_training/NORM_NORM_1.csv'
df2 = pd.read_csv(file2)
x2_v2 = df2['V2']
x2_v8 = df2['V8']
x2_v10 = df2['V10']

plt.figure(figsize=(10,8))
# plt.plot(time, x2_v2, '-r', label = 'Sensor_NORM', )
plt.plot(time, x2_v8, '-r', label = 'ChargeAmp_NORM')
plt.plot(time, x2_v10, '-g', label = 'Filter_NORM')

# plt.plot(time, x1_v2, '--r', label = 'Sensor_Fault', )
plt.plot(time, x1_v8, '--r', label = 'ChargeAmp_Fault')
plt.plot(time, x1_v10, '--g', label = 'Filter_Fault')

plt.xlim(0, 0.0015)
plt.xlabel('Time', fontsize = 13)
plt.ylabel('Normalized Signal', fontsize = 13)

plt.title('NORM vs AMP(C1) Fault', fontsize = 20)
plt.legend()
```
