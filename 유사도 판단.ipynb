```python
import numpy as np
import os
import pandas as pd
from fastdtw import fastdtw


data_fold = './Data/data_training'


db = {}

for filename in os.listdir(data_fold):
    if filename.endswith('.csv'):
        file_path = os.path.join(data_fold, filename)
        db[filename] = pd.read_csv(file_path)
        
print(list(db.keys())[0])

fk = list(db.keys())
sk = list(db.keys())

for first_db_key in fk:
    sk.remove(first_db_key)
    for second_db_key in sk:
        v2_d, path = fastdtw(db[first_db_key]['V2'], db[second_db_key]['V2'])
        v8_d, path = fastdtw(db[first_db_key]['V8'], db[second_db_key]['V8'])
        v10_d, path = fastdtw(db[first_db_key]['V10'], db[second_db_key]['V10'])
        
        if v2_d + v8_d + v10_d < 100:
            print(f'{first_db_key} - {second_db_key} : V2 [{v2_d:10.4f}] | V2 [{v8_d:10.4f}] | V2 [{v10_d:10.4f}] | TOT {v2_d + v8_d + v10_d}')
        
print('Done')
        
        
        
```
